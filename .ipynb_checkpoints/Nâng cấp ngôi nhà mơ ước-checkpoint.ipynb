```python
#[Bài tập] Xây nhà ước mơ sử dụng hàm
import turtle
t = turtle.Turtle()

t.goto(-120,0)
t.fd(270)
t.lt(90)
t.fd(90)
t.lt(90)
t.fd(270)
t.lt(90)
t.fd(90)

t.goto(5,0)
t.rt(180)
for i in range(4):
    t.fd(40)
    t.rt(90)
    
t.pu()    
t.goto(-120,90)
t.pd()
t.rt(30)
t.fd(120)
t.rt(90)
t.goto(90,90)

t.color("blue","bisque")
t.begin_fill()
t.pu()
t.goto(-200,-50)
t.lt(120)
t.pd()
t.goto(-205,0)
t.rt(90)
t.goto(-230,0)
t.rt(90)
t.goto(-250,-50)
t.rt(90)
t.goto(-200,-50)
t.end_fill()

def ve_cay_thong(a,x,y):
    t.color("blue","green")
    t.begin_fill()
    for i in range(3):
        t.fd(a)
        t.rt(120)
    t.end_fill()
    t.pu()
    t.goto(x,a*((3**(1/2))/2)+y)
    t.pd()
    t.begin_fill()
    for i in range(3):
           t.fd(((a/2)-20)*2)
           t.rt(120)
    t.end_fill()
    b = ((a/2)-20)*2
    c = a*((3**(1/2))/2)
    d = b*((3**(1/2))/2)
    t.pu()
    t.goto(x-20,c+d+y)
    t.pd()
    t.begin_fill()
    for i in range(3):
        t.fd(((b/2)-20)*2)
        t.rt(120)
    t.end_fill()

t.pu()
t.goto(-150,0)
t.pd()
ve_cay_thong(120,-170,0)

t.pu()
t.goto(250,-50)
t.pd()
t.rt(90)

t.color("blue","bisque")
t.begin_fill()
t.goto(255,-30)
t.rt(90)
t.goto(270,-30)
t.rt(90)
t.goto(280,-50)
t.rt(90)
t.goto(250,-50)
t.end_fill()

t.pu()
t.goto(300,-30)
t.pd()
ve_cay_thong(90,280,-30)


    


```
