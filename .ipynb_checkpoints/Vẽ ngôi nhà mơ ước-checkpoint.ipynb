```python
import turtle

t = turtle.Turtle()

t.penup()
t.bk(200)
t.pendown()

for i in range(4) :
    t.fd(120)
    t.lt(90)
t.fd(40)
t.lt(90)

t.fd(60)
t.rt(90)
t.fd(40)
t.rt(90)
t.fd(60)

t.lt(90)
t.fd(40)

for i in range(2) :
    t.lt(45)
    t.fd(100)
    t.lt(45)
    t.fd(120)
    t.lt(90)

t.lt(70)
t.penup()
t.fd(102)
t.rt(70)
t.pendown()

for i in range(4) :
    t.fd(20)
    t.lt(90)

t.lt(70)
t.penup()
t.fd(102)
t.rt(70)
t.lt(135)
t.pendown()

for i in range(3) :
    t.fd(100)
    t.lt(90)

t.penup()
t.lt(135)
t.fd(120)
t.pendown()

t.rt(125)
t.fd(100)







    
```
