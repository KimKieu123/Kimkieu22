```python
#vẽ ngôi nhà mơ ước
from turtle import *
bgcolor('turquoise')
speed(0)
pu()
goto(-300, -300)
pd()


def tm(x):
	fillcolor(x)
	
begin_fill()
for i in range(4): 
	tm('blue')
	fd(180)
	lt(90)
end_fill()

fd(180/3)
lt(90)

begin_fill()
for i in range(2): 
	tm('green')
	fd(180/2)
	rt(90)
	fd(180/3)
	rt(90)
end_fill()

lt(90)
fd(180/3)
rt(90)
fd(180)


begin_fill()
tm('pink')
b = (180*(2**0.5)/2)
rt(45)
fd(b)
rt(90)
fd(b)
end_fill()


begin_fill()
tm('orange')
for i in range(4):
	lt(90)
	fd(b)
end_fill()


begin_fill()
tm('yellow')
lt(90)
fd(b)
rt(135)
fd(180)
rt(45)
fd(b)
end_fill()

bk(b/3)
rt(135)
pu()
fd(180/2)
pd()


begin_fill()
tm('brown')
for i in range(2):
	fd180/4)
	rt(45)
	fd(b/3)
	rt(135)
end_fill()

pu()
bk(180/2)
rt(90)
fd(180*1.7)
pd()


begin_fill()
tm('brown')
for i in range(2):
	fd(60/5)
	lt(90)
	fd(c/2)
	lt(90)
end_fill()

lt(90)
fd(c/2)
rt(90)
bk(60/2.5)


tm('green')
for i in range(3):
	begin_fill()
	d = 60 * (2**0.5) / 2
	lt(45)
	fd(d)
	rt(90)
	fd(d)
	rt(135)
	fd(60)
	rt(90)
	end_fill()
	pu()
	fd(60/2)
	rt(90)
	fd(c*0.05)
	pd()
	c *= 0.9

pu()
lt(90)
fd(180)
pd()

e = 180/5

begin_fill()
tm('yellow')
circle(e)
end_fill()
for i in range(4):
	rt(90)
	fd(e)
	bk(e)
	lt(90)
	circle(e, 45)
	rt(90)
	fd(e/2)
	bk(e/2)
	lt(90)
	circle(e, 45)

done()
```
